```python
print("CALCULADORA DE IMC")

lista = []
soma_imc = 0

# contadores
abaixo = 0
normal = 0
excesso = 0
ob1 = 0
ob2 = 0
ob3 = 0

while True:
    nome = input("Qual seu nome?: ")
    peso = float(input("Qual seu peso?: "))
    altura = float(input("Qual sua altura?: "))

    imc = peso / (altura * altura)

    # classificaÃ§Ã£o
    if imc < 18.5:
        classificacao = "Abaixo do peso"
        abaixo += 1

    elif imc <= 24.9:
        classificacao = "Peso normal"
        normal += 1

    elif imc <= 29.9:
        classificacao = "Excesso de peso"
        excesso += 1

    elif imc <= 34.9:
        classificacao = "Obesidade classe I"
        ob1 += 1

    elif imc <= 39.9:
        classificacao = "Obesidade classe II"
        ob2 += 1

    else:
        classificacao = "Obesidade classe III"
        ob3 += 1

    lista.append([nome, imc, classificacao])

    soma_imc += imc

    continuar = input("Deseja adicionar outra pessoa? (s/n): ")

    if continuar.lower() == "n":
        break

# organiza do maior IMC para o menor
lista.sort(key=lambda pessoa: pessoa[1], reverse=True)

print("\n===== RANKING DE IMC =====")

for pessoa in lista:
    print(
        "Nome:", pessoa[0],
        "| IMC:", round(pessoa[1], 2),
        "| ClassificaÃ§Ã£o:", pessoa[2]
    )

# mÃ©dia dos IMCs
media = soma_imc / len(lista)

print("\nMÃ©dia dos IMCs:", round(media, 2))

# total de pessoas
total = len(lista)

# porcentagens
p_abaixo = (abaixo / total) * 100
p_normal = (normal / total) * 100
p_excesso = (excesso / total) * 100
p_ob1 = (ob1 / total) * 100
p_ob2 = (ob2 / total) * 100
p_ob3 = (ob3 / total) * 100

print("\n===== PORCENTAGEM DOS ALUNOS =====")

print(round(p_abaixo, 1), "% dos alunos estÃ£o abaixo do peso")
print(round(p_normal, 1), "% dos alunos estÃ£o no peso normal")
print(round(p_excesso, 1), "% dos alunos estÃ£o com excesso de peso")
print(round(p_ob1, 1), "% dos alunos estÃ£o com obesidade classe I")
print(round(p_ob2, 1), "% dos alunos estÃ£o com obesidade classe II")
print(round(p_ob3, 1), "% dos alunos estÃ£o com obesidade classe III")
```
